# Lahingu modelleerimine
Olgu kaks armeed üksteise vastu lahingus. Kuidas saaks matemaatiliselt ennustada, kumb armee lahingu võidab?

Olgu $x(t)$ ja $y(t)$ vastavalt esimese ja teise armee tugevus ajahetkel $t$. On keeruline arvuliselt määrata, mis on armee tugevus, sest tuleb arvesse võtta sõdurite arvu, sõjalist valmisolekut, relvajõude, ülema kvaliteeti jne. Ülesande lihtsustamiseks ütleme, et tugevus on võrdne sõdurite arvuga.

Eeldame, et $x(t)$ ja $y(t)$ varieeruvad pidevalt ja diferentsiaalselt ajafunktsioonidena. See on idealiseerimine, sest tugevus peab reaalsuses olema alati täisarv (armees ei saa olla 10000,2 sõdurit). Samas, võib vaielda, et kui tugevus on suur, siis ühe või kahe võrra seda muuta on juba nii väike muutus, et võib ka sama hästi muuta seda väiksemate kogustega. Näiteks, kui armeel on kell 8 70000 sõdurit ja kell 9 69995 sõdurit, tundub mõistlik määrata kell 8:05 sõdurite arvuks 69999,5833. Samuti, teeme need funktsioonid diferentseeruvaks, et nende graafikud n-ö siledaks teha.

Kuigi meil puudub kindel valem $x(t)$ määramiseks, on meil see-eest hea kogus infot armee operatsioonilise kaotuse (OK) kohta (s.t. kaotused, mis tulenevad haiguste, deserteerimise ja teiste lahinguväliste asjade tõttu), lahinguliste kaotuste kohta (LK) ning abivägede kohta (AV). Seega, saame armee tugevuse muutuse valemi kujul
$$\frac{dx(t)}{dt}=-(OK+LK)+AV.$$
Sarnane valem kehtib ka $y(t)$ kohta. Probleem on leida asjakohased valemid, et määrata, kes lahingu võidab.

Eeldame, et tegu on konventsionaalse lahinguga, s.t. kaks armeed avatud maastikul üksteise vastu. Siis avaldub lahingmudel kujul
\begin{align*}
\frac{dx(t)}{dt}&=-ax(t)-by(t)+P(t),\\
\frac{dy(t)}{dt}&=-cx(t)-dy(t)+Q(t),
\end{align*}
kus $a$ ja $d$ on vastavate armeede operatsiooniliste kaotuste kordajad, $b$ ja $c$ lahinguliste kaotuste kordajad ning $P(t)$ ja $Q(t)$ abivägede suurus päeva kohta. Kui armee on isoleeritud, siis abivägede suurus oleks konstantne 0. Eeldatavasti oleks abivägede suurus mittenegatiivne, aga samas sõdureid võidakse lahingust ka eemaldada. Operatsiooniliste kaotuste määrad $-ax(t)$ ja $-dy(t)$ määravad konstantse kaotuste määra (kui lahingutegevust ei toimu ja abivägesi pole oodata:
$$\frac{dx/dt}{x}=-a,\quad\frac{dy/dt}{y}=-d.$$
Teisi liiki kaotusi on ka võimalikud, aga praegused annavad küllaltki hea ülevaate enamustele situatsioonidele.

Kui ainukesed määrad on abiväed ja operatsioonilised kaotused, siis ei ole mingisugust lahingutegevust, sest kummalgi armeel pole oma vastase vastu mingit effekti. Lahingukaotuste määrad $-by(t)$ ja $-cx(t)$ toob sisse tegeliku lahingtegevuse. Konventsionaalne armee tegeleb avatud maastikul ning eeldame, et iga armee liige on vastase surmatsoonis. Samuti eeldame, et kui armee liige saab kannatada, siis vastase tuli suunatakse kohe teiste sõdurite poole. Nende tingimuste all on $-by(t)$ lahingkaotuste määr, kus $b$ on vastase armee lahinguefektiivsuse kordaja. 

Kuidas arvutada armee lahinguefektiivsust? Üks viis oleks seada
$$b=r_yp_y,$$
ku $r_y$ on vastase armee tulistamiskiirus ja $p_y$ on tõenäosus, et üks vastase lask tapab oma sõduri. Nende määramine on võimalik ainult pärast lahingut, kui on lahingust küllaltki andmeid. Samas, kui on andmeid varasemast lahingust, siis saab ka lihtsustamise mõttes neid kasutada, et tulistamiskiirus ja tapmistõenäosust ennustada.

## Lihtsustatud mudeli läbimängimine
Olgu kaks konventsionaalset armeed, kes peavad üksteisega lahingut ideaalsetes konditsioonides, s.t. operatsioonilisi kaotusi ei teki ja abivägesid juurde ei kutsuta. Sel juhul avaldub diferentsiaalvõrrandisüsteem kujul
\begin{align*}
\frac{dx}{dt}&=-by,\\
\frac{dy}{dt}&=-cx.
\end{align*}
Jagades omavahel need võrrandid, saame
$$\frac{dy}{dx}=\frac{cx}{by}.$$
Muutujate eraldades ja integreerides saame
$$b\int_{y_0}^{y(t)}y\,dy=c\int_{x_0}^{x(t)}x\,dx$$
$$b(y^2(t)-y_0^2)=c(x^2(t)-x_0^2).$$
Olgu $K$ konstant $by_0^2-cx_0^2$. Võrrandi
$$by^2-cx^2=K$$
graafik on parabool (või kaks sirget, kui $K=0$). Esitame need graafikul erinevate $K$ väärtuste korral.

```python
import numpy as np
import matplotlib.pyplot as plt

b=0.1
c=0.3
K=0

def f(x):
    return np.sqrt((K+c*x**2)/b)

x = np.linspace(0, 10, 400)
plt.plot(x, f(x), label = 'K=0')
plt.arrow(2, f(2), -0.03, -f(2.03)+f(2), shape='full', lw=0, length_includes_head=True, head_width=.3)
K=-1
plt.plot(x, f(x), label = 'K=-1')
plt.arrow(3, f(3), -0.03, -f(3.03)+f(3), shape='full', lw=0, length_includes_head=True, head_width=.3)
K=1
plt.plot(x, f(x), label = 'K=1')
plt.arrow(2, f(2), -0.03, -f(2.03)+f(2), shape='full', lw=0, length_includes_head=True, head_width=.3)
K=2
plt.plot(x, f(x), label = 'K=2')
plt.arrow(2, f(2), -0.03, -f(2.03)+f(2), shape='full', lw=0, length_includes_head=True, head_width=.3)
K=-2
plt.plot(x, f(x), label = 'K=-2')
plt.arrow(3, f(3), -0.03, -f(3.03)+f(3), shape='full', lw=0, length_includes_head=True, head_width=.3)
plt.legend()
plt.xlim(0, 5)
plt.ylim(0, 10)
plt.xlabel('x(t)')
plt.ylabel('y(t)')
plt.show()
```

On selge, et mõistlik on vaadata ainult esimest veerandit. Nooled joonte peal näitavad armee tugevuse muutumise suunda ajas.

Kes võidab sellises lahingus? Ütleme, et üks armee võidab, kui teine armee on täielikult hävitatud (tema tugevus on 0). Näiteks, $y$ võidab, kui $K>0$, sest võrrandi $by^2+cx^2=K$ korral y ei saavuta kunagi väärtust 0, samas $x$ on hävitatud, kui $y(t)$ saavutab väärtuse $\sqrt{K/b}$. Seega, $y$ armee soovib saavutada olukorra, kus $K>0$, s.t. nad tahavad, et kehtiks võrratus
$$by_0^2>cx_0^2.$$

Võrrand $by^2-cx^2=K$ näitab ainult kahe armee tugevuste sõltuvust üksteisest, aga mitte aja kulgemist. Seda saame näidata, kui võtame algse võrrandisüsteemi
\begin{align*}
\frac{dx}{dt}&=-by,\\
\frac{dy}{dt}&=-cx,
\end{align*}
leiame esimese võrrandi tuletise
$$\frac{d^2x}{dt^2}=-b\frac{dy}{dt}$$
ja asendame sinna teise võrrandi
$$\frac{d^2x}{dt^2}=bcx.$$
Kasutades algtingimusteks
$$x_0=x(0),\quad -by_0=\frac{dx}{dt}\vert_{t=0},$$
saame diferentsiaalvõrrandi lahenduseks
$$x(t)=x_0\cosh \beta t-\gamma y_0\sinh\beta t,$$
kus $\beta=\sqrt{bc}$ ja $\gamma=\sqrt{b/c}$. y armee võrrand on analoogiliselt
$$y(t)=y_0\cosh\beta t-\frac{x_0}{\gamma}\sinh\beta t.$$
Joonestame nende funktsioonide graafikud juhul, kui $K>0$ (s.t. $\gamma y_0>x_0$).

```python
b=0.03
c=0.01
beta=np.sqrt(b*c)
gamma=np.sqrt(b/c)
x_0=100
y_0=70

def f1(x):
    return x_0*np.cosh(beta*x)-gamma*y_0*np.sinh(beta*x)

def f2(x):
    return y_0*np.cosh(beta*x)-x_0/gamma*np.sinh(beta*x)

def f3(x):
    return gamma*y_0-x_0

x=np.linspace(0,70,400)
plt.plot(x, f1(x), label='x(t)')
plt.plot(x, f2(x), label='y(t)')
plt.xlim(0, 68)
plt.ylim(0, 100)
plt.legend()
plt.xlabel('aeg')
plt.ylabel('tugevus')
plt.show()
```

Paneme tähele, et ei ole vajalik, et y armee algne suurus oleks x armeest suurem, vaid piisab sellest, et $\gamma y_0>x_0$. Analoogiliselt, kui $\gamma y_0<x_0$, siis mudeli põhjal võidaks lahingu x armee, ning kui $\gamma y_0=x_0$, siis jääksid armeed viiki (mis praktikas on väga ebatõenäoline).
